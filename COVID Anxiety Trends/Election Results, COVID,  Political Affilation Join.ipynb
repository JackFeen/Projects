# COVID-19 Cases + Search Trends by Political Affilation Join

This code changes the data types and aggregations of our datasets so they can be properly joined.

```python
import pandas as pd
```

```python
search_df = pd.read_csv("/Users/jackfeen/search_trends_w_political_affilation.csv")
search_df.head()
```

```python
covid_df = pd.read_csv("/Users/jackfeen/covid_cases.csv")
covid_df.head()
```

```python
search_df['date'] = pd.to_datetime(search_df['date'])

# 2. Create a new column for the week (this will aggregate dates into weeks)
search_df['week'] = search_df['date'].dt.to_period('W')

# 3. Group by 'state' and 'week', then aggregate the numeric columns (mean)
aggregated_df = search_df.groupby(['Party Affiliation', 'week']).agg({
    'anxiety': 'mean',   # Replace 'value1' with your actual column name
   
}).reset_index()
aggregated_df.head(30)
```

```python
covid_df['start_date'] = pd.to_datetime(covid_df['start_date'])

# 2. Create a new column for the week (this will aggregate dates into weeks)
covid_df['week'] = covid_df['start_date'].dt.to_period('W')

# 3. Group by 'state' and 'week', then aggregate the numeric columns (mean)
ag_cov_df = covid_df.groupby(['week']).agg({
    'new_cases': 'sum',   # Replace 'value1' with your actual column name
   
}).reset_index()
ag_cov_df.head(30)
```

```python
merged_df = pd.merge(aggregated_df, ag_cov_df, on='week', how='inner')
merged_df.head()
```

```python
merged_df.to_csv("")
```
